##### Import Necessary Modules:

## Tasks


<h2>1. Load the Dataset</h2>
<p>
We use the <code>pandas.read_csv()</code> function for reading CSV files. However, in this version of the lab, which operates on JupyterLite, the dataset needs to be downloaded to the interface using the provided code below.

The functions below will download the dataset into your browser:



```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### 2. Explore the Dataset
##### Task 1: Display basic information and summary statistics of the dataset.


```python
## Write your code here
df.info()
df.describe()
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here

missing_values = df.isnull().sum()

# Print the count of missing values for each column
print("Missing values for each column:")
print(missing_values)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
# Visualize missing values using a heatmap
plt.figure(figsize=(12, 8))  # Set the figure size

# Create the heatmap
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False, xticklabels=df.columns)

# Show the plot
plt.title('Missing Values Heatmap')
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here

# Count the number of missing rows for the 'Employment' column
missing_employment = df['Employment'].isnull().sum()

# Print the result
print(f"Number of missing rows in 'Employment' column: {missing_employment}")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here

# Identify the most frequent value in the 'Employment' column
most_frequent_employment = df['Employment'].mode()[0]

# Print the most frequent value
print(f"The most frequent value in 'Employment' column is: {most_frequent_employment}")
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here

# Impute missing values in the 'Employment' column with the most frequent value
df['Employment'] = df['Employment'].fillna(most_frequent_employment)

# Verify that missing values have been imputed
missing_values_after_imputation = df['Employment'].isnull().sum()
print(f"Missing values in 'Employment' column after imputation: {missing_values_after_imputation}")

# Save the cleaned dataset
df.to_csv("cleaned_survey_data_with_imputed_employment.csv", index=False)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here

# Create a bar chart to visualize the distribution of the 'Employment' column
employment_counts = df['Employment'].value_counts()

# Create the bar chart with a single color
plt.figure(figsize=(10, 6))  # Set the figure size

# Plot the bar chart
plt.barh(employment_counts.index, employment_counts.values, color='skyblue')  # Use a simple color

# Add title and labels
plt.title('Distribution of Employment Status After Imputation')
plt.xlabel('Employment Status')
plt.ylabel('Frequency')

# Show the plot
plt.tight_layout()  
plt.show()
```
